```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
```

```python
"""
Balanced Data: made from notebook: Copy of Chunks - BERT
Highlights Data: Made from notebook: Stage2_Chunks
"""
train_sentence_data_path = '/content/drive/MyDrive/266 Final Project/Full Orchestration: Stage 1+2/final_data/Old Data/sentence_training_data.csv'
test_sentence_data_path = '/content/drive/MyDrive/266 Final Project/Full Orchestration: Stage 1+2/final_data/Old Data/sentence_testing_data.csv'
```

```python
train_sentence_data = pd.read_csv(train_sentence_data_path)
test_sentence_data = pd.read_csv(test_sentence_data_path)
```

```python
train_sentence_data
```

```python
# Split the data into features (sentences) and labels
X_train = train_sentence_data['doc_text'].values
y_train = train_sentence_data['label'].values
X_test = test_sentence_data['doc_text'].values
y_test = test_sentence_data['label'].values
```

```python
# Vectorize the sentences using TF-IDF
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
# Initialize and train the classifier
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
